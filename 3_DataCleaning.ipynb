With data transformation we are changing the value of the dataset, transforming the data according to some rules we defined.

We may be interested to substitute the value of a null with the value of the row above:

```python
data.fillna(method='ffill') # fill NaN/Null using the value of the record above
```

Why we need to substitute null values? There are some application like ML that do not want to have null values, so they should be substituted with some values.

There are two options if the null value are not so numeros:

* (Null treatment technique) delete the rows in which you have the null. With many tuples and few null values this does not affect the result of the ML technique.
* (data transformation technique) substitute the null value with another value, e.g. the value of the tuple above.

```python
data.fillna(method='bfill') # to substitute even the Nan/Null at the top of the table
```

## Split and drop columns

We notice that the Name and Surname are in the same column, so we want to split the attribute, since a **good design require atomic values**.

```python
data2.Name.str.split(expand=True) # the expand tell to create as many columns as split output
```

The latter command is not so good, indeed in case of multiple names it can create more than two column as we desire.

```python
data2[['First', 'Last']] = data2.Name.str.split(pat=' ', n=1, expand=True) # where pat is the delimiter for the split and n is the number of delimiter it will split on
data2
```

Obviously now we added two columns that contain the same information of the Name column, in a better scheme format, but we have redundancy. So we have to drop the column Name.

```python
data2.drop(['Name'], axis=1) # axis=1 stays for drop column
# equivalently we can do
#data2.drop(columns=['Name])
```

#### An example using n=1 for the split delimiter search

```python
df = pd.DataFrame({
    'Address': ['Via Ponzio 34, Milano, Lombardia', 'Via Golgi 39, Milano, Lombardia', 'Viale Romagna5, Roma, Lazio']
})
df
```

## Format

```python
df2 = pd.DataFrame(
    data={'Currency': {0: 30, 1: 1.34, 2: 100.3456},
          'Int': {0: 34, 1: 46, 2: 345},
          'Rate': {0: 0.30, 1: 0.98, 2: 0.0456},
    }
)
df2
```

We may want to apply a specific format:    
* a certain number of decimals
* currency symbol
* percentage symbol

```python
format_mapping = {'Currency': '${:,.2f}', 
                  'Int':'{:.0f}', 
                  'Rate': '{:,.2f}%'}
df2.style.format(format_mapping)
```
